```python
final_list = []
for i in job_list:
    final_list.append("%20".join(i.split(" "))) 
```

```python
# Job counts in companies based on the category

BASE_URL = "https://api.adzuna.com/v1/api"
count2 = []
salary =[]
job_category1 = []
for i in final_list:
    hist_data=requests.get("https://api.adzuna.com/v1/api/jobs/us/histogram?app_id=3cbdefa2&app_key=e1a2e1a4a3c5712842b5bdc3fcf6cf2b&what={}".format(i))
    histjson = hist_data.json()
    for j in list(histjson['histogram'].keys()):
        salary.append(j)
        count2.append(histjson['histogram'][j])
        job_category1.append(" ".join(i.split("%20")))
```

```python
histogram_data = pd.DataFrame(columns = ['Approximate Salary','Count of Jobs in that salary range','Category'])
histogram_data['Approximate Salary'] = salary
histogram_data['Count of Jobs in that salary range'] = count2
histogram_data['Category'] = job_category1
```

```python
# Job counts in companies based on the category

BASE_URL = "https://api.adzuna.com/v1/api"
company = []
count1 = []
job_category = []
for i in final_list:
    data=requests.get("https://api.adzuna.com/v1/api/jobs/us/top_companies?app_id=3cbdefa2&app_key=e1a2e1a4a3c5712842b5bdc3fcf6cf2b&what={}".format(i))
    datajson = data.json()
    for j in datajson['leaderboard']:
        company.append(j['canonical_name'])
        count1.append(j['count'])
        #salary.append(j['average_salary'])
        job_category.append(" ".join(i.split("%20")))
    
```

```python
company_data = pd.DataFrame(columns = ['Company','Count of Jobs in that company','Category'])
company_data['Company'] = company
company_data['Count of Jobs in that company'] = count1
company_data['Category'] = job_category
```

```python
geodata = requests.get("https://api.adzuna.com/v1/api/jobs/us/geodata?app_id=3cbdefa2&app_key=e1a2e1a4a3c5712842b5bdc3fcf6cf2b")
```

```python
geojson = geodata.json()
```

```python
area = []
count = []
for i in geojson['locations']:
    area.append(i['location']['display_name'])
    count.append(i['count'])
```

```python
job_locations = pd.DataFrame(columns = ['Location','Count of Jobs at that location'])
```
